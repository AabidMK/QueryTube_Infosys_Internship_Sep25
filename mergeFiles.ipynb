```python
import pandas as pd

# Read the CSV and Excel files
csv_df = pd.read_csv('Dhruvin_cleaned_video_details_final.csv')
xls_df = pd.read_excel('Dhruvin_cleaned_transcripts.xls')  # Changed to .xls

# Merge on the 'id' column
merged_df = pd.merge(csv_df, xls_df, on='id', how='inner')  # Use 'outer', 'left', or 'right' if needed

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_output.csv', index=False)

# Display the first few rows
merged_df.head()
```
