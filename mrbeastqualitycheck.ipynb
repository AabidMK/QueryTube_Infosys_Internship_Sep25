## After **Cleaning**

```python
import pandas as pd

transcripts_df = pd.read_csv("/content/mrbeasttranscripts.csv")
channel_details_df = pd.read_csv("/content/youtube channel details.csv")

print("===== TRANSCRIPTS DATA =====")
print("Shape:", transcripts_df.shape)
print("\nInfo:")
print(transcripts_df.info())
print("\nMissing values:\n", transcripts_df.isnull().sum())
print("\nSample:\n", transcripts_df.head())

print("\n===== CHANNEL DETAILS DATA =====")
print("Shape:", channel_details_df.shape)
print("\nInfo:")
print(channel_details_df.info())
print("\nMissing values:\n", channel_details_df.isnull().sum())
print("\nSample:\n", channel_details_df.head())

channel_details_df["description"] = channel_details_df["description"].fillna("No description available")

if "tags" in channel_details_df.columns:
    channel_details_df = channel_details_df.drop(columns=["tags"])
    print("\n'Dropped tags column (all values missing)'")

print("\nAfter Cleaning - Missing values:\n", channel_details_df.isnull().sum())

print("\nNumerical Statistics:")
print(channel_details_df.describe())
```
